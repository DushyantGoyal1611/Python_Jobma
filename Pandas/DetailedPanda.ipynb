```python
import numpy as np
```

```python
df = pd.read_csv('nba.csv')
df
```

```python
# from here i will learn pandas in full detail
```

```python
# Access a single column
df['Team'].head()
```

```python
# Access multiple columns
df[['Name','Team','Age']].head()
```

```python
# Create new Column
df['Married'] = 'Unknown'
```

```python
df.head()
```

```python
# delete multiple columns
df.drop(['Married','Weight'],axis=1,inplace=True)
df.head()
```

```python
# rename the column
df.rename(columns={'Name':'PlayerName'},inplace=True)
df.head()
```

```python
# rename multiple columns
df.rename(columns={'PlayerName':'Name','Number':'Num'},inplace=True)
df.head()
```

```python
df.columns
```

```python
# # to check how many null values are there in column
df.isnull().sum()
```

```python
# Now we play with rows
```

```python
# Rows selection using loc[]

num = df.loc[0,'Age']
print(num)
```

```python
# Select Multiple rows and columns

num = df.loc[0:4,['Name','Team','Age']]
num
```

```python
# all rows and multiple columns

num = df.loc[:,['Name','Team','Age']]
num
```

```python
# Boolean Indexing -> Filters rows based on condition that evaluates to a boolean(True or False)
filtered_data = df[df['Age']%2==0]
filtered_data.head()
```

```python
# Multiple Conditions

filtered_data = df[(df['Age']%2==0) & (df['Age']>22)]
filtered_data.head()
```

```python
# Aggregate Functions 
```

```python
df.aggregate({'Num':['sum','min'],
              'Age':['sum','max'],
              'Name':['count'],
              'Team': ['count']
            })
```

```python
# Groupby
```

```python
group1 = df.groupby('Name')
print(group1.groups)
```

```python
print(group1.first().head())
```

```python
# Grouping with multiple keys

group2 = df.groupby(['Name','Position'])
print(group2.groups)
```

```python
print(group2.first().head())
```

```python
# Iterating Grouping
for name,group in group1:
    print(name)
    print(group)
    print()
```

```python
# Selecting a group

group2.get_group(('Aaron Brooks','PG'))
```

```python
# Applying functions on Groups
```

```python
#Aggregation
result = group1['Age'].aggregate(np.sum)
print(result.head())
```

```python
result = group1['Age'].aggregate([np.sum,np.mean,np.min])
print(result)
```

```python
# Transformation
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
group3 = df.groupby('Name')
```

```python
df['Age'] = group3['Age'].transform(lambda x : x.fillna(x.mean()))
```

```python
df.isnull().sum()
```

```python
# Filtration
result = group3.filter(lambda x: x['Age'].max()>34)
print(result)
```
