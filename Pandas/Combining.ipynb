```python
# Merging, Joining, and Concatenating
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
# reading the csv files

df1 = pd.read_csv('nba.csv')
df2 = pd.read_csv('titanic.csv')
```

```python
# Concatenating

frames = [df1,df2]
res = pd.concat(frames)
print(res)
```

```python
# Merging

res = pd.merge(df1,df2, on='Age')
res.head()
```

```python
res = pd.merge(df1,df2, on=['Age','Name'])
res.head()
```

```python
res = pd.merge(df1,df2,how='left',on=['Age'])
res
```

```python
res = pd.merge(df1,df2,how='right',on=['Age'])
res
```

```python
res = pd.merge(df1,df2,how='inner',on=['Age'])
res
```

```python
res = pd.merge(df1,df2,how='outer',on=['Age'])
res
```
