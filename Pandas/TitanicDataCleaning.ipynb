```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df
```

```python
df.set_index('PassengerId',inplace=True)
```

```python
df
```

```python
df.columns
```

```python
df.rename(columns={'SibSp':'Siblings/Spouse','ParCh':'Parent/Child','Sex':'Gender'},inplace=True)
df
```

```python
if pd.isnull(df.loc[921,'Gender']):
    df.loc[921,'Gender'] = 'male'
```

```python
df.isnull().sum()
```

```python
mean_value = df['Survived'].mean()
df['Survived'].fillna(mean_value,inplace = True)
```

```python
df.isnull().sum()
```

```python
df.loc[df['Fare'].idxmax()]
df.loc[df['Fare'].idxmin()]
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Fare'] = df.groupby('Pclass')['Fare'].transform(lambda x : x.fillna(x.mean()))
```

```python
df.isnull().sum()
```

```python
print(df['Fare'].isna().sum())
```

```python
if pd.isnull(df['Embarked']).any():
    df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
df.loc[1]
```

```python
#delete the "Cabin" Column
df.drop('Cabin',axis=1,inplace=True)
```

```python
df
```

```python
mean_value = df['Age'].mean()
df['Age'].fillna(mean_value,inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
df[['Survived','Age']] = df[['Survived','Age']].astype(int)
```

```python
df
```

```python
#Now visualizing the data using matplotlib
```

```python
xpoints = df['Gender']
ypoints1 = df['Survived']

plt.plot(xpoints,ypoints1,marker='o',ms=5,mec='r',mfc='r',ls='dashed',lw=3,color='hotpink')

plt.suptitle("My Project")  #super title
plt.title("Line Chart",loc='center')
plt.xlabel("Gender")
plt.ylabel("Survived")
plt.grid()

plt.show()
```

```python
xpoints = df['Gender']
ypoints = df['Name']

plt.xlabel("Gender")
plt.ylabel("Name")

plt.plot(xpoints,ypoints,marker='o',mec='r',mfc='r',ls='dotted')

plt.show()
```

```python
x  = df[df['Survived']==1]['Gender'].value_counts()
labels = ['Men','Women']
colors = ['red','hotpink']
plt.pie(x,labels=labels,colors=colors,startangle=90)
plt.legend()

```

```python
x  = df[df['Survived']==1]['Gender'].value_counts()

labels = ['Men','Women']
colors = ['red','hotpink']
explodes=[0,0.2]
plt.suptitle("My Project")
plt.title("Pie Chart")

plt.pie(x,labels=labels,colors=colors,explode=explodes,startangle=90)
plt.legend(loc='best')

plt.show()
```
