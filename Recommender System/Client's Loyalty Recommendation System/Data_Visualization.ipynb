```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data_collection.csv')
```

```python
df.shape
```

```python
df.drop(['subscription_type_x',
         'premium_storage_y',
         'currency_val',
         'jobma_catcher_sub_accounts',
         'jobma_job_company_profile'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df['plan_type'].fillna('No', inplace=True)
```

```python
df['is_unlimited'].fillna('No', inplace=True)
```

```python
for col in df.columns:
    if df[col].dtype == np.int64:
        df[col].fillna(0, inplace=True)
    elif df[col].dtype == np.float64:
        df[col].fillna(0.0, inplace=True)
    elif df[col].dtype == object:
        df[col].fillna('Unknown', inplace=True)
```

# Visualization

**HeatMap**

```python
corr_matrix = df.corr(numeric_only=True)
plt.figure(figsize=(20, 18))
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu', linewidths=0.2, linecolor='gray', square=True)
plt.title('Relation between Columns')
plt.tight_layout()
plt.show()
```
